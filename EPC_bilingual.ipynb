# Medio Poroso

```python
def create_heterogeneous_tight_rock(shape, porosity_target=0.10, voxel_size_um=0.05, seed=None):
    """
    Genera un medio poroso tipo "bundle of tubes" con capilares rectos paralelos.
    
    Características:
    - Capilares rectos (sin tortuosidad) como en modelos clásicos de Poiseuille
    - Distribución regular con variación en diámetros
    - Capilares paralelos que atraviesan el dominio
    - Heterogeneidad controlada en los diámetros
    """
    
    print(f"="*60)
    print(f"GENERANDO MEDIO POROSO TIPO HAZ DE CAPILARES")
    print(f"="*60)
    print(f"Dimensiones: {shape}")
    print(f"Porosidad objetivo: {porosity_target:.1%}")
    print(f"Tamaño de voxel: {voxel_size_um:.3f} μm")
    
    if seed is not None:
        np.random.seed(seed)
    
    # PASO 1: Definir distribución de diámetros de capilares
    # Distribución heterogénea pero organizada
    
    # Escala 1: Capilares grandes (20% del total)
    large_pore_size_um = 0.5  # Diámetro grande
    fraction_large = 0.20
    
    # Escala 2: Capilares medianos (50% del total)
    medium_pore_size_um = 0.3  # Diámetro medio
    fraction_medium = 0.50
    
    # Escala 3: Capilares pequeños (30% del total)
    small_pore_size_um = 0.15  # Diámetro pequeño
    fraction_small = 0.30
    
    # Inicializar medio
    medium = np.zeros(shape, dtype=bool)
    
    # PASO 2: Calcular número y espaciado de capilares
    # Dirección principal de los capilares (eje X por defecto)
    flow_direction = 2  # 0=Z, 1=Y, 2=X
    
    if flow_direction == 2:  # Capilares en dirección X
        cross_section = shape[0] * shape[1]
        length = shape[2]
        n_y, n_z = shape[1], shape[0]
    elif flow_direction == 1:  # Capilares en dirección Y
        cross_section = shape[0] * shape[2]
        length = shape[1]
        n_x, n_z = shape[2], shape[0]
    else:  # Capilares en dirección Z
        cross_section = shape[1] * shape[2]
        length = shape[0]
        n_x, n_y = shape[2], shape[1]
    
    # Estimar número de capilares basado en porosidad objetivo
    avg_radius_um = (large_pore_size_um * fraction_large + 
                     medium_pore_size_um * fraction_medium + 
                     small_pore_size_um * fraction_small) / 2
    avg_radius_vox = avg_radius_um / voxel_size_um
    avg_area_per_capillary = np.pi * avg_radius_vox**2
    
    total_pore_area_needed = cross_section * porosity_target
    n_capillaries = int(total_pore_area_needed / avg_area_per_capillary)
    
    # Distribuir capilares por tamaño
    n_large_pores = max(2, int(n_capillaries * fraction_large))
    n_medium_pores = max(5, int(n_capillaries * fraction_medium))
    n_small_pores = max(5, int(n_capillaries * fraction_small))
    
    print(f"\nDistribución de capilares rectos:")
    print(f"• Grandes (Ø {large_pore_size_um:.2f} μm): {n_large_pores} capilares")
    print(f"• Medianos (Ø {medium_pore_size_um:.2f} μm): {n_medium_pores} capilares")
    print(f"• Pequeños (Ø {small_pore_size_um:.2f} μm): {n_small_pores} capilares")
    print(f"• Total: {n_large_pores + n_medium_pores + n_small_pores} capilares")
    
    # PASO 3: Función para crear capilares rectos
    def create_straight_capillary(position, diameter_um, direction='x'):
        """
        Crea un capilar recto a través del dominio.
        
        Args:
            position: (y, z) para dirección X, (x, z) para Y, o (x, y) para Z
            diameter_um: Diámetro del capilar
            direction: 'x', 'y', o 'z'
        """
        radius_vox = diameter_um / (2 * voxel_size_um)
        
        if direction == 'x':
            y_center, z_center = position
            for x in range(shape[2]):
                for y in range(max(0, int(y_center - radius_vox - 1)), 
                              min(shape[1], int(y_center + radius_vox + 2))):
                    for z in range(max(0, int(z_center - radius_vox - 1)), 
                                  min(shape[0], int(z_center + radius_vox + 2))):
                        if (y - y_center)**2 + (z - z_center)**2 <= radius_vox**2:
                            medium[z, y, x] = True
        
        elif direction == 'y':
            x_center, z_center = position
            for y in range(shape[1]):
                for x in range(max(0, int(x_center - radius_vox - 1)), 
                              min(shape[2], int(x_center + radius_vox + 2))):
                    for z in range(max(0, int(z_center - radius_vox - 1)), 
                                  min(shape[0], int(z_center + radius_vox + 2))):
                        if (x - x_center)**2 + (z - z_center)**2 <= radius_vox**2:
                            medium[z, y, x] = True
        
        else:  # direction == 'z'
            x_center, y_center = position
            for z in range(shape[0]):
                for x in range(max(0, int(x_center - radius_vox - 1)), 
                              min(shape[2], int(x_center + radius_vox + 2))):
                    for y in range(max(0, int(y_center - radius_vox - 1)), 
                                  min(shape[1], int(y_center + radius_vox + 2))):
                        if (x - x_center)**2 + (y - y_center)**2 <= radius_vox**2:
                            medium[z, y, x] = True
    
    # PASO 4: Generar posiciones de capilares con distribución semi-regular
    print(f"\nGenerando capilares rectos paralelos...")
    
    # Lista para almacenar todas las posiciones y tamaños
    capillary_specs = []
    
    # Generar grid semi-regular
    if flow_direction == 2:  # X direction
        # Estimar espaciado basado en número de capilares
        n_grid = int(np.sqrt(n_large_pores + n_medium_pores + n_small_pores))
        dy = shape[1] / (n_grid + 1)
        dz = shape[0] / (n_grid + 1)
        
        # Capilares grandes - distribución más espaciada
        for i in range(n_large_pores):
            # Posición base en grid
            grid_y = (i % int(np.sqrt(n_large_pores)) + 1) * shape[1] / (int(np.sqrt(n_large_pores)) + 2)
            grid_z = (i // int(np.sqrt(n_large_pores)) + 1) * shape[0] / (int(np.sqrt(n_large_pores)) + 2)
            
            # Añadir variación aleatoria
            y_pos = grid_y + np.random.uniform(-dy/4, dy/4)
            z_pos = grid_z + np.random.uniform(-dz/4, dz/4)
            
            # Asegurar margen
            margin = large_pore_size_um / (2 * voxel_size_um) + 1
            y_pos = np.clip(y_pos, margin, shape[1] - margin)
            z_pos = np.clip(z_pos, margin, shape[0] - margin)
            
            capillary_specs.append(((y_pos, z_pos), large_pore_size_um, 'large'))
    
    # Función para verificar solapamiento
    def check_overlap(new_pos, new_radius, existing_specs):
        for (pos, diam, _) in existing_specs:
            dist = np.sqrt((new_pos[0] - pos[0])**2 + (new_pos[1] - pos[1])**2)
            min_dist = (new_radius + diam/2) / voxel_size_um + 1
            if dist < min_dist:
                return True
        return False
    
    # Capilares medianos - llenar espacios
    attempts = 0
    added_medium = 0
    while added_medium < n_medium_pores and attempts < n_medium_pores * 10:
        if flow_direction == 2:
            y_pos = np.random.uniform(0.1, 0.9) * shape[1]
            z_pos = np.random.uniform(0.1, 0.9) * shape[0]
            pos = (y_pos, z_pos)
        
        radius = medium_pore_size_um / 2
        
        if not check_overlap(pos, radius, capillary_specs):
            capillary_specs.append((pos, medium_pore_size_um, 'medium'))
            added_medium += 1
        attempts += 1
    
    # Capilares pequeños - llenar espacios restantes
    attempts = 0
    added_small = 0
    while added_small < n_small_pores and attempts < n_small_pores * 10:
        if flow_direction == 2:
            y_pos = np.random.uniform(0.05, 0.95) * shape[1]
            z_pos = np.random.uniform(0.05, 0.95) * shape[0]
            pos = (y_pos, z_pos)
        
        radius = small_pore_size_um / 2
        
        if not check_overlap(pos, radius, capillary_specs):
            capillary_specs.append((pos, small_pore_size_um, 'small'))
            added_small += 1
        attempts += 1
    
    # PASO 5: Crear todos los capilares
    direction_map = {0: 'z', 1: 'y', 2: 'x'}
    flow_dir = direction_map[flow_direction]
    
    print(f"  Creando {len(capillary_specs)} capilares en dirección {flow_dir.upper()}...")
    
    for i, (position, diameter, size_class) in enumerate(capillary_specs):
        # Añadir pequeña variación en el diámetro para heterogeneidad
        diameter_variation = np.random.uniform(0.9, 1.1)
        actual_diameter = diameter * diameter_variation
        
        create_straight_capillary(position, actual_diameter, flow_dir)
        
        if (i + 1) % 50 == 0:
            print(f"    Progreso: {i + 1}/{len(capillary_specs)} capilares creados")
    
    # PASO 6: Análisis final
    final_porosity = np.sum(medium) / medium.size
    
    # Verificar conectividad
    inlet = np.zeros(shape, dtype=bool)
    if flow_direction == 2:  # X
        inlet[:, :, :5] = True
    elif flow_direction == 1:  # Y
        inlet[:, :5, :] = True
    else:  # Z
        inlet[:5, :, :] = True
    
    connected = ps.filters.trim_disconnected_blobs(im=medium, inlets=inlet)
    connectivity = np.sum(connected) / np.sum(medium) if np.sum(medium) > 0 else 0
    
    # Calcular distribución de diámetros
    dt_final = distance_transform_edt(medium)
    pore_radii = dt_final[medium] * voxel_size_um * 2  # Convertir a diámetros
    heterogeneity = np.std(pore_radii) / np.mean(pore_radii) if len(pore_radii) > 0 else 0
    
    print(f"\nResultados finales:")
    print(f"• Porosidad final: {final_porosity:.3f}")
    print(f"• Conectividad: {connectivity:.3f}")
    print(f"• Heterogeneidad: {heterogeneity:.3f}")
    print(f"• Número total de capilares: {len(capillary_specs)}")
    
    if len(pore_radii) > 0:
        print(f"• Diámetro medio efectivo: {np.mean(pore_radii):.3f} μm")
        print(f"• Rango de diámetros: {np.min(pore_radii):.3f} - {np.max(pore_radii):.3f} μm")
    
    # Tortuosidad = 1.0 para capilares rectos
    print(f"• Tortuosidad: 1.00 (capilares rectos)")
    
    # Información adicional sobre la distribución
    n_large_actual = sum(1 for _, _, t in capillary_specs if t == 'large')
    n_medium_actual = sum(1 for _, _, t in capillary_specs if t == 'medium')
    n_small_actual = sum(1 for _, _, t in capillary_specs if t == 'small')
    
    print(f"\nDistribución final de capilares:")
    print(f"• Grandes: {n_large_actual} ({n_large_actual/len(capillary_specs)*100:.1f}%)")
    print(f"• Medianos: {n_medium_actual} ({n_medium_actual/len(capillary_specs)*100:.1f}%)")
    print(f"• Pequeños: {n_small_actual} ({n_small_actual/len(capillary_specs)*100:.1f}%)")
    
    domain_size_um = [dim * voxel_size_um for dim in shape]
    
    return {
        'medium': medium,
        'shape': shape,
        'voxel_size_um': voxel_size_um,
        'domain_size_um': domain_size_um,
        'porosity_target': porosity_target,
        'porosity_actual': final_porosity,
        'connectivity': connectivity,
        'heterogeneity': heterogeneity,
        'pore_scales': {
            'large': large_pore_size_um,
            'medium': medium_pore_size_um,
            'small': small_pore_size_um
        }
    }
```

# Funciones Euler-Poincaré Caracteristico

```python
# ========== FUNCIONES DE ANÁLISIS EPC ==========

def calculate_euler_characteristic_robust(binary_image):
    """
    Método robusto para calcular la característica de Euler.
    χ = β₀ - β₁ + β₂
    """
    
    print(f"  {L('calculating_epc_robust')}")
    
    struct_elem_26 = ndimage.generate_binary_structure(3, 3)
    labeled, n_components = ndimage.label(binary_image, structure=struct_elem_26)
    
    struct_elem_6 = ndimage.generate_binary_structure(3, 1)
    inverted = ~binary_image
    labeled_inv, n_cavities_total = ndimage.label(inverted, structure=struct_elem_6)
    
    border_labels = set()
    for label_val in range(1, n_cavities_total + 1):
        mask = labeled_inv == label_val
        if (np.any(mask[0, :, :]) or np.any(mask[-1, :, :]) or 
            np.any(mask[:, 0, :]) or np.any(mask[:, -1, :]) or
            np.any(mask[:, :, 0]) or np.any(mask[:, :, -1])):
            border_labels.add(label_val)
    
    n_cavities = n_cavities_total - len(border_labels)
    n_cavities = max(0, n_cavities)
    
    try:
        from skimage.measure import euler_number
        euler_char = euler_number(binary_image, connectivity=3)
        
    except ImportError:
        n_voxels = np.sum(binary_image)
        
        if n_voxels > 0:
            kernel = np.ones((3, 3, 3))
            kernel[1, 1, 1] = 0
            neighbors = ndimage.convolve(binary_image.astype(int), kernel, mode='constant')
            total_connections = np.sum(neighbors[binary_image]) // 2
            
            expected_tree_connections = max(0, n_voxels - n_components)
            excess_connections = max(0, total_connections - expected_tree_connections)
            
            density = n_voxels / binary_image.size
            adjustment_factor = 1.0 + density
            
            n_loops = int(excess_connections / (26 * adjustment_factor))
            
            if n_components == 1 and density > 0.05:
                n_loops += int(n_voxels * density * 0.001)
            
        else:
            n_loops = 0
        
        euler_char = n_components - n_loops + n_cavities
    
    beta_1 = n_components + n_cavities - euler_char
    
    if beta_1 < 0:
        print(f"    ⚠️ β₁ < 0, ajustando...")
        beta_1 = max(0, n_components - 1)
        euler_char = n_components - beta_1 + n_cavities
    
    volume = np.sum(binary_image)
    euler_density = euler_char / volume if volume > 0 else 0
    
    print(f"    {L('components')} (β₀): {n_components}")
    print(f"    {L('tunnels_loops')} (β₁): {beta_1}")
    print(f"    {L('cavities')} (β₂): {n_cavities}")
    print(f"    χ = β₀ - β₁ + β₂ = {n_components} - {beta_1} + {n_cavities} = {euler_char}")
    print(f"    χ/V: {euler_density:.6f}")
    
    if abs(euler_char) > 1000:
        print(f"    ⚠️ {L('warning_chi_range')}")
    elif euler_char < 0:
        print(f"    ✓ {L('chi_negative_high_conn')}")
    elif euler_char > 0:
        print(f"    ✓ {L('chi_positive_low_conn')}")
    else:
        print(f"    ✓ {L('chi_near_zero')}")
    
    return euler_char, euler_density, n_components

calculate_euler_characteristic_local_config = calculate_euler_characteristic_robust


def calculate_connectivity_function(medium, voxel_size_um=0.05, n_points=30):
    """
    Calcula la función de conectividad χ(d) variando el diámetro de poro.
    """
    
    print("\n" + "="*60)
    print(L('calculating_epc'))
    print("="*60)
    
    dt = distance_transform_edt(medium)
    max_radius_vox = np.max(dt)
    max_radius_um = max_radius_vox * voxel_size_um
    
    print(f"{L('max_radius_3d')}: {max_radius_um:.2f} μm")
    print(f"{L('total_volume')}: {medium.shape}")
    print(f"{L('porous_voxels')}: {np.sum(medium):,}")
    
    euler_initial, euler_dens_initial, beta_0_initial = calculate_euler_characteristic_robust(medium)
    print(f"\n{L('initial_epc')}: χ = {euler_initial} (β₀ = {beta_0_initial})")
    
    if euler_initial > 0:
        print(f"⚠️ {L('warning_initial_chi_positive')}")
        print(f"   - {L('poorly_connected')}")
        print(f"   - {L('many_isolated')}")
        print(f"   - {L('isolated_pore_structure')}")
    else:
        print(f"✓ {L('initial_chi_negative')}")
        print(f"  {L('magnitude_suggests').format(abs(euler_initial))}")
    
    radii_um = np.concatenate([[0], np.linspace(0.01, max_radius_um * 0.9, n_points-1)])
    
    euler_chars = []
    euler_densities = []
    porosities = []
    n_components = []
    
    print(f"\n{L('calculating_epc_radii')}")
    print(f"{L('radius_header')} | χ | β₀ | {L('porosity_header')} | {L('comment_header')}")
    print("-" * 60)
    
    euler, euler_dens, beta_0 = euler_initial, euler_dens_initial, beta_0_initial
    euler_chars.append(euler)
    euler_densities.append(euler_dens)
    porosities.append(np.sum(medium) / medium.size)
    n_components.append(beta_0)
    
    print(f"{0:9.3f} | {euler:6d} | {beta_0:4d} | {porosities[0]:9.4f} | {L('initial_state')}")
    
    opened_images = [medium.copy()]
    
    for i, radius in enumerate(radii_um[1:], 1):
        if radius < voxel_size_um:
            opened = medium.copy()
        else:
            radius_vox = int(radius / voxel_size_um)
            if radius_vox < 1:
                opened = medium.copy()
            else:
                struct = morphology.ball(radius_vox)
                opened = morphology.binary_opening(medium, struct)
        
        opened_images.append(opened)
        
        if np.sum(opened) > 0:
            euler, euler_dens, beta_0 = calculate_euler_characteristic_robust(opened)
            porosity = np.sum(opened) / opened.size
        else:
            euler, euler_dens, beta_0 = 0, 0, 0
            porosity = 0
        
        euler_chars.append(euler)
        euler_densities.append(euler_dens)
        porosities.append(porosity)
        n_components.append(beta_0)
        
        if i == 1 or i % 5 == 0 or i == len(radii_um)-1:
            comment = ""
            if euler > 0 and euler_chars[i-1] <= 0:
                comment = L('transition_low_conn')
            elif beta_0 > n_components[i-1] * 2:
                comment = L('significant_fragmentation')
            
            print(f"{radius:9.2f} | {euler:6d} | {beta_0:4d} | {porosity:9.4f} | {comment}")
    
    radii_um = np.array(radii_um)
    euler_chars = np.array(euler_chars)
    euler_densities = np.array(euler_densities)
    porosities = np.array(porosities)
    n_components = np.array(n_components)
    
    print(f"\n{L('chi_evolution_analysis')}")
    
    dchi = np.diff(euler_chars)
    if np.all(dchi >= 0):
        print(f"✓ {L('chi_monotonic')}")
    else:
        n_decreases = np.sum(dchi < 0)
        print(f"⚠ {L('chi_not_monotonic').format(n_decreases)}")
    
    critical_radius = None
    sign_changes = np.where(np.diff(np.sign(euler_chars)))[0]
    
    if len(sign_changes) > 0:
        for idx in sign_changes:
            if euler_chars[idx] < 0 and euler_chars[idx+1] > 0:
                r1, r2 = radii_um[idx], radii_um[idx+1]
                chi1, chi2 = euler_chars[idx], euler_chars[idx+1]
                
                critical_radius = r1 - chi1 * (r2 - r1) / (chi2 - chi1)
                
                print(f"\n✓ {L('critical_radius_found').format(critical_radius)}")
                print(f"  {L('transition').format(r1, chi1, r2, chi2)}")
                print(f"  {L('at_rc_percolation')}")
                break
        
        if critical_radius is None:
            critical_radius = radii_um[sign_changes[0]]
            print(f"\n⚠ {L('atypical_crossing').format(critical_radius)}")
    else:
        min_idx = np.argmin(np.abs(euler_chars))
        critical_radius = radii_um[min_idx]
        min_chi = euler_chars[min_idx]
        
        if min_chi < 0:
            print(f"\n⚠ {L('no_zero_crossing')}")
            print(f"  {L('chi_remains_negative').format(abs(min_chi), critical_radius)}")
            print(f"  {L('medium_maintains_conn')}")
        else:
            print(f"\n⚠ {L('chi_always_positive').format(min_chi, critical_radius)}")
            print(f"  {L('medium_low_initial_conn')}")
    
    print(f"\n{L('topological_analysis_summary')}")
    print(f"• {L('chi_range')}: [{min(euler_chars)}, {max(euler_chars)}]")
    print(f"• {L('components')}: {n_components[0]} → {n_components[-1]}")
    print(f"• {L('porosity_reduction')}: {porosities[0]:.1%} → {porosities[-1]:.1%}")
    
    if euler_chars[0] < -100:
        print(f"• {L('medium_type')}: {L('highly_connected')}")
    elif euler_chars[0] < 0:
        print(f"• {L('medium_type')}: {L('well_connected')}")
    elif euler_chars[0] < 100:
        print(f"• {L('medium_type')}: {L('moderate_connectivity')}")
    else:
        print(f"• {L('medium_type')}: {L('poorly_connected_isolated')}")
    
    if max(np.abs(euler_chars)) > 1000:
        print(f"\n⚠️ {L('warning_chi_out_of_range')}")
        print(f"   {L('may_indicate_error')}")
    
    return {
        'radii': radii_um,
        'euler_chars': euler_chars,
        'euler_densities': euler_densities,
        'porosities': porosities,
        'n_components': n_components,
        'critical_radius': critical_radius,
        'opened_images': opened_images
    }


def calculate_connectivity_function_enhanced(medium, voxel_size_um=0.05, n_points=30):
    """
    Versión mejorada con detección de cambios abruptos.
    """
    
    epc_results = calculate_connectivity_function(medium, voxel_size_um, n_points)
    
    critical_radius_physical, abrupt_changes = detect_abrupt_changes_in_connectivity(
        epc_results['radii'],
        epc_results['euler_chars'],
        epc_results['n_components'],
        epc_results['porosities'],
        voxel_size_um
    )
    
    epc_results['critical_radius_physical'] = critical_radius_physical
    epc_results['critical_radius_mathematical'] = epc_results['critical_radius']
    epc_results['abrupt_changes_analysis'] = abrupt_changes
    
    if critical_radius_physical and epc_results['critical_radius']:
        print(f"\n{L('critical_radii_comparison')}")
        print(f"• {L('physical_rc_abrupt')}: {critical_radius_physical:.3f} μm")
        print(f"• {L('mathematical_rc')}: {epc_results['critical_radius']:.3f} μm")
        print(f"• {L('difference')}: {abs(epc_results['critical_radius'] - critical_radius_physical):.3f} μm")
        
        if critical_radius_physical < epc_results['critical_radius']:
            print(f"✓ {L('physical_before_expected')}")
        else:
            print(f"⚠️ {L('mathematical_before_check')}")
    
    return epc_results
```

# $R_c$ detection

```python
def detect_abrupt_changes_in_connectivity(radii, euler_chars, n_components, porosities, voxel_size_um):
    """
    Detecta cambios abruptos en la función de conectividad χ(r).
    """
    
    print("\n" + "="*60)
    print(L('abrupt_changes_analysis'))
    print("="*60)
    
    n_pts = len(radii)
    
    dchi_dr = np.gradient(euler_chars, radii)
    
    window_length = min(7, n_pts if n_pts % 2 == 1 else n_pts - 1)
    if window_length >= 3:
        dchi_dr_smooth = savgol_filter(dchi_dr, window_length=window_length, polyorder=2)
    else:
        dchi_dr_smooth = dchi_dr
    
    d2chi_dr2 = np.gradient(dchi_dr_smooth, radii)
    
    sign_changes = np.where(np.diff(np.sign(euler_chars)))[0]
    
    positive_derivs = dchi_dr_smooth[dchi_dr_smooth > 0]
    if len(positive_derivs) > 0:
        threshold = np.percentile(positive_derivs, 75)
    else:
        threshold = 0
    
    maxima, max_properties = find_peaks(dchi_dr_smooth, 
                                       height=threshold,
                                       prominence=np.std(dchi_dr_smooth) * 0.5)
    
    minima, min_properties = find_peaks(-dchi_dr_smooth,
                                       prominence=np.std(dchi_dr_smooth) * 0.5)
    
    log_components = np.log10(n_components + 1)
    d_log_comp = np.gradient(log_components, radii)
    
    critical_radii_candidates = []
    
    # Método 1: Máximo en dχ/dr
    if len(maxima) > 0:
        heights = dchi_dr_smooth[maxima]
        sorted_indices = np.argsort(heights)[::-1]
        
        primary_max_idx = maxima[sorted_indices[0]]
        r_primary_max = radii[primary_max_idx]
        
        critical_radii_candidates.append((L('method_max_loss'), r_primary_max, primary_max_idx))
        
        print(f"\nMétodo 1 - {L('method_max_loss')} (máximo en dχ/dr):")
        print(f"  {L('radius')}: {r_primary_max:.3f} μm")
        print(f"  dχ/dr: {dchi_dr_smooth[primary_max_idx]:.2f}")
        print(f"  χ: {euler_chars[primary_max_idx]}")
        print(f"  {L('components')}: {n_components[primary_max_idx]}")
    
    # Método 2: χ cruza cero
    if len(sign_changes) > 0:
        for sc_idx in sign_changes:
            if euler_chars[sc_idx] < 0 and euler_chars[sc_idx + 1] >= 0:
                r1, r2 = radii[sc_idx], radii[sc_idx + 1]
                chi1, chi2 = euler_chars[sc_idx], euler_chars[sc_idx + 1]
                r_zero = r1 - chi1 * (r2 - r1) / (chi2 - chi1)
                
                critical_radii_candidates.append((L('method_topological'), r_zero, sc_idx))
                
                print(f"\nMétodo 2 - {L('method_topological')}:")
                print(f"  {L('radius')}: {r_zero:.3f} μm")
                break
    
    # Método 3: Fragmentación
    comp_peaks, _ = find_peaks(d_log_comp, prominence=0.1)
    if len(comp_peaks) > 0:
        max_comp_change_idx = comp_peaks[np.argmax(d_log_comp[comp_peaks])]
        r_comp_change = radii[max_comp_change_idx]
        critical_radii_candidates.append((L('method_fragmentation'), r_comp_change, max_comp_change_idx))
        
        print(f"\nMétodo 3 - {L('method_fragmentation')}:")
        print(f"  {L('radius')}: {r_comp_change:.3f} μm")
    
    # Método 4: Puntos de inflexión
    inflection_points = np.where(np.diff(np.sign(d2chi_dr2)))[0]
    if len(inflection_points) > 0:
        for ip in inflection_points:
            if ip > 0 and ip < len(dchi_dr_smooth) - 1:
                if dchi_dr_smooth[ip] > threshold:
                    r_inflection = radii[ip]
                    critical_radii_candidates.append((L('method_inflection'), r_inflection, ip))
                    
                    print(f"\nMétodo 4 - {L('method_inflection')}:")
                    print(f"  {L('radius')}: {r_inflection:.3f} μm")
                    break
    
    # Método 5: Pérdida de porosidad
    if len(porosities) > 0 and porosities[0] > 0:
        porosity_ratio = porosities / porosities[0]
        
        porosity_drop_idx = np.where(porosity_ratio < 0.5)[0]
        if len(porosity_drop_idx) > 0:
            idx_50 = porosity_drop_idx[0]
            r_porosity_50 = radii[idx_50]
            critical_radii_candidates.append((L('method_porosity_loss'), r_porosity_50, idx_50))
            
            print(f"\nMétodo 5 - {L('method_porosity_loss')}:")
            print(f"  {L('radius')}: {r_porosity_50:.3f} μm")
    
    # Seleccionar radio crítico
    if critical_radii_candidates:
        max_deriv_candidates = [c for c in critical_radii_candidates if L('method_max_loss') in c[0]]
        
        if max_deriv_candidates:
            primary_method, primary_radius, primary_idx = max_deriv_candidates[0]
        else:
            critical_radii_candidates.sort(key=lambda x: x[1])
            primary_method, primary_radius, primary_idx = critical_radii_candidates[0]
        
        print(f"\n" + "="*50)
        print(L('physical_critical_radius'))
        print(f"  {L('method')}: {primary_method}")
        print(f"  {L('radius')}: {primary_radius:.3f} μm")
        print(f"  χ: {euler_chars[min(primary_idx, len(euler_chars)-1)]:.0f}")
        print(f"  β₀: {n_components[min(primary_idx, len(n_components)-1)]}")
        print(f"  {L('porosity')}: {porosities[min(primary_idx, len(porosities)-1)]:.3f}")
        
        if len(critical_radii_candidates) >= 3:
            confidence = L('high')
        elif len(critical_radii_candidates) >= 2:
            confidence = L('medium_conf')
        else:
            confidence = L('low')
        print(f"  {L('confidence')}: {confidence} ({len(critical_radii_candidates)} {L('matching_methods')})")
        print("="*50)
        
        results = {
            'primary_radius': primary_radius,
            'primary_method': primary_method,
            'primary_idx': primary_idx,
            'all_candidates': critical_radii_candidates,
            'dchi_dr': dchi_dr,
            'dchi_dr_smooth': dchi_dr_smooth,
            'd2chi_dr2': d2chi_dr2,
            'maxima': maxima,
            'minima': minima,
            'max_properties': max_properties,
            'confidence': confidence
        }
        
        return primary_radius, results
    else:
        print(f"\n⚠️ {L('no_abrupt_changes')}")
        return None, None
```

```python
def create_epc_static_comparison(epc_results):
    """
    Crea una imagen estática mostrando estados clave de la evolución EPC.
    Con soporte bilingüe.
    """
    
    FONT_SIZE_MAIN = 20
    FONT_SIZE_TITLE = 22
    FONT_SIZE_SUPTITLE = 24
    FONT_SIZE_SMALL = 16
    FONT_SIZE_LEGEND = 18
    
    plt.rcParams.update({
        'font.family': 'sans-serif',
        'font.serif': ['Sans-serif'],
        'font.size': FONT_SIZE_MAIN,
        'axes.labelsize': FONT_SIZE_MAIN,
        'axes.titlesize': FONT_SIZE_TITLE,
        'xtick.labelsize': FONT_SIZE_MAIN,
        'ytick.labelsize': FONT_SIZE_MAIN,
        'legend.fontsize': FONT_SIZE_LEGEND,
        'figure.titlesize': FONT_SIZE_SUPTITLE,
    })
    
    voxel_size_um = epc_results.get('voxel_size_um', 0.05)
    domain_size_um = epc_results.get('domain_size_um', None)
    
    if domain_size_um is None:
        if 'shape' in epc_results:
            shape = epc_results['shape']
        else:
            shape = epc_results['opened_images'][0].shape
        domain_size_um = [dim * voxel_size_um for dim in shape]
    
    print(f"{L('creating_static')}: {domain_size_um[0]:.1f} × {domain_size_um[1]:.1f} × {domain_size_um[2]:.1f} μm³")
    
    n_total = len(epc_results['radii'])
    indices = [0, n_total//3, 2*n_total//3, n_total-1]
    
    fig = plt.figure(figsize=(20, 12))
    gs = fig.add_gridspec(2, 4, height_ratios=[1.5, 1], hspace=0.3, wspace=0.3)
    
    step = 3
    scale_factor = voxel_size_um * step
    
    for i, idx in enumerate(indices):
        ax = fig.add_subplot(gs[0, i], projection='3d')
        
        radius = epc_results['radii'][idx]
        opened = epc_results['opened_images'][idx]
        opened_vis = opened[::step, ::step, ::step]
        
        porosity_real = np.sum(opened) / opened.size
        
        if np.sum(opened_vis) > 10:
            try:
                verts, faces, _, _ = measure.marching_cubes(opened_vis.astype(float), level=0.5)
                verts_um = verts * scale_factor
                
                if epc_results['euler_densities'][idx] < 0:
                    color = plt.cm.Blues(0.7)
                else:
                    color = plt.cm.Reds(0.7)
                
                ax.plot_trisurf(verts_um[:, 0], verts_um[:, 1], verts_um[:, 2],
                               triangles=faces, color=color, alpha=0.8,
                               edgecolor='none')
            except:
                ax.text(0.5, 0.5, 0.5, L('no_data'), transform=ax.transAxes, 
                       ha='center', va='center', fontsize=FONT_SIZE_SMALL,
                       bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))
        else:
            ax.text(0.5, 0.5, 0.5, L('no_connected_structure'), transform=ax.transAxes, 
                   ha='center', va='center', fontsize=FONT_SIZE_SMALL,
                   bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))
        
        ax.set_xlim([0, domain_size_um[2]])
        ax.set_ylim([0, domain_size_um[1]])
        ax.set_zlim([0, domain_size_um[0]])
        
        ax.set_xlabel(L('x_axis'), fontsize=FONT_SIZE_MAIN)
        ax.set_ylabel(L('y_axis'), fontsize=FONT_SIZE_MAIN)
        ax.set_zlabel(L('z_axis'), fontsize=FONT_SIZE_MAIN, rotation=0, ha='right')
        
        ax.tick_params(axis='x', labelsize=FONT_SIZE_SMALL)
        ax.tick_params(axis='y', labelsize=FONT_SIZE_SMALL)
        ax.tick_params(axis='z', labelsize=FONT_SIZE_SMALL)
        
        ax.xaxis.set_major_locator(plt.MaxNLocator(4))
        ax.yaxis.set_major_locator(plt.MaxNLocator(4))
        ax.zaxis.set_major_locator(plt.MaxNLocator(4))
        
        ax.set_title( 
            f'r = {radius:.2f} μm\n'
            f'χ = {epc_results["euler_chars"][idx]:.0f}\n'
            f'$\phi$ = {porosity_real * 100:.1f}%',
            fontsize=FONT_SIZE_MAIN,
            weight='normal'
        )

        ax.view_init(elev=30, azim=45)
        ax.grid(True, alpha=0.3)
    
    # Panel inferior: Función de conectividad
    ax_bottom = fig.add_subplot(gs[1, :])
    
    ax_bottom.plot(epc_results['radii'], epc_results['euler_chars'], 
                  'b-', linewidth=2.5)
    ax_bottom.axhline(y=0, color='r', linestyle='--', alpha=0.7, linewidth=1.5)
    
    if 'critical_radius' in epc_results and epc_results['critical_radius'] is not None:
        ax_bottom.axvline(x=epc_results['critical_radius'], color='g', 
                         linestyle='--', linewidth=2,
                         label=f'{L("rc_topological")} = {epc_results["critical_radius"]:.2f} μm')
    
    if 'critical_radius_physical' in epc_results and epc_results['critical_radius_physical'] is not None:
        ax_bottom.axvline(x=epc_results['critical_radius_physical'], color='red', 
                         linestyle=':', linewidth=2,
                         label=f'{L("rc_derivative")} = {epc_results["critical_radius_physical"]:.2f} μm')
    
    for idx in indices:
        ax_bottom.plot(epc_results['radii'][idx], epc_results['euler_chars'][idx], 
                      'ro', markersize=10)
    
    #ax_bottom.fill_between(epc_results['radii'], 0, epc_results['euler_chars'],
     #                     where=(epc_results['euler_chars'] < 0),
      #                    color='blue', alpha=0.1, label=L('legend_high_conn'))
    ax_bottom.fill_between(epc_results['radii'], 0, epc_results['euler_chars'],
                          where=(epc_results['euler_chars'] > 0),
                          color='red', alpha=0.1, label=L('legend_low_conn'))
    
    ax_bottom.set_xlabel(L('pore_radius'), fontsize=FONT_SIZE_MAIN)
    ax_bottom.set_ylabel(L('euler_char'), fontsize=FONT_SIZE_MAIN)
    
    ax_bottom.grid(True, alpha=0.3)
    ax_bottom.legend(fontsize=FONT_SIZE_LEGEND, loc='best', ncol=2)
    ax_bottom.tick_params(axis='both', which='major', labelsize=FONT_SIZE_MAIN)
    
    plt.tight_layout(rect=[0, 0.03, 1, 0.97])
    
    filename = 'epc_evolution_3d_static.png'
    plt.savefig(filename, dpi=300, bbox_inches='tight', facecolor='white', edgecolor='none')
    print(f"{L('static_saved')}: {filename}")
    print(f"{L('visualized_domain')}: {domain_size_um[0]:.1f} × {domain_size_um[1]:.1f} × {domain_size_um[2]:.1f} μm³")
    plt.close(fig)
```

# EPC ANIMATION

```python
def create_animated_epc_evolution(epc_results, fps=2):
    """
    Crea una animación GIF mostrando la evolución de la conectividad.
    Con soporte bilingüe.
    """
    
    FONT_SIZE_MAIN = 20
    FONT_SIZE_TITLE = 22
    FONT_SIZE_SUPTITLE = 24
    FONT_SIZE_SMALL = 16
    FONT_SIZE_LEGEND = 18
    
    plt.rcParams.update({
        'font.family': 'sans-serif',
        'font.serif': ["Sans-serif"],
        'font.size': FONT_SIZE_MAIN,
        'axes.labelsize': FONT_SIZE_MAIN,
        'axes.titlesize': FONT_SIZE_TITLE,
        'xtick.labelsize': FONT_SIZE_MAIN,
        'ytick.labelsize': FONT_SIZE_MAIN,
        'legend.fontsize': FONT_SIZE_LEGEND,
        'figure.titlesize': FONT_SIZE_SUPTITLE,
    })
    
    print(f"\n{L('creating_3d_animation')}")
    
    voxel_size_um = epc_results.get('voxel_size_um', 0.05)
    domain_size_um = epc_results.get('domain_size_um', None)
    
    if domain_size_um is None:
        if 'shape' in epc_results:
            shape = epc_results['shape']
        else:
            shape = epc_results['opened_images'][0].shape
        domain_size_um = [dim * voxel_size_um for dim in shape]
    
    print(f"{L('domain_for_animation')}: {domain_size_um[0]:.1f} × {domain_size_um[1]:.1f} × {domain_size_um[2]:.1f} μm³")
    
    n_frames = min(20, len(epc_results['radii']))
    frame_indices = np.linspace(0, len(epc_results['radii'])-1, n_frames, dtype=int)
    
    fig = plt.figure(figsize=(16, 8), facecolor='white')
    
    step = 3
    scale_factor = voxel_size_um * step
    
    euler_min = min(epc_results['euler_chars'])
    euler_max = max(epc_results['euler_chars'])
    
    if euler_min > 0:
        euler_min = -abs(euler_max) * 0.1
    if euler_max < 0:
        euler_max = abs(euler_min) * 0.1
    
    print(f"{L('chi_range_for_vis')}: [{euler_min:.0f}, {euler_max:.0f}]")
    
    def update(frame_idx):
        fig.clear()
        
        idx = frame_indices[frame_idx]
        radius = epc_results['radii'][idx]
        
        ax1 = fig.add_subplot(121, projection='3d')
        ax2 = fig.add_subplot(122)
        
        opened = epc_results['opened_images'][idx]
        opened_vis = opened[::step, ::step, ::step]
        
        porosity_real = np.sum(opened) / opened.size
        
        n_components = epc_results['n_components'][idx]
        euler_char = epc_results['euler_chars'][idx]
        
        if np.sum(opened_vis) > 10:
            try:
                verts, faces, _, _ = measure.marching_cubes(opened_vis.astype(float), level=0.5)
                verts_um = verts * scale_factor
                
                if euler_char < 0:
                    color_map = plt.cm.Blues(0.7)
                else:
                    color_map = plt.cm.Reds(0.7)
                
                surf = ax1.plot_trisurf(verts_um[:, 0], verts_um[:, 1], verts_um[:, 2],
                                       triangles=faces, 
                                       color=color_map,
                                       alpha=0.85,
                                       edgecolor='none',
                                       linewidth=0,
                                       shade=True)
                
            except Exception as e:
                coords = np.where(opened_vis)
                if len(coords[0]) > 0:
                    n_points = min(5000, len(coords[0]))
                    indices = np.random.choice(len(coords[0]), n_points, replace=False)
                    
                    x_um = coords[2][indices] * scale_factor
                    y_um = coords[1][indices] * scale_factor
                    z_um = coords[0][indices] * scale_factor
                    
                    ax1.scatter(x_um, y_um, z_um, c='black', s=0.5, alpha=0.3)
        else:
            ax1.text(0.5, 0.5, 0.5, L('no_connected_structure'), 
                    transform=ax1.transAxes, ha='center', va='center',
                    fontsize=FONT_SIZE_SMALL, 
                    bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))
        
        ax1.set_xlim([0, domain_size_um[2]])
        ax1.set_ylim([0, domain_size_um[1]])
        ax1.set_zlim([0, domain_size_um[0]])
        
        ax1.set_xlabel(L('x_axis'), fontsize=FONT_SIZE_MAIN, labelpad=10)
        ax1.set_ylabel(L('y_axis'), fontsize=FONT_SIZE_MAIN, labelpad=10)
        ax1.set_zlabel(L('z_axis'), fontsize=FONT_SIZE_MAIN, rotation=0, ha='right', labelpad=10)
        
        ax1.tick_params(axis='x', labelsize=FONT_SIZE_SMALL)
        ax1.tick_params(axis='y', labelsize=FONT_SIZE_SMALL)
        ax1.tick_params(axis='z', labelsize=FONT_SIZE_SMALL)
        
        ax1.xaxis.set_major_locator(plt.MaxNLocator(5))
        ax1.yaxis.set_major_locator(plt.MaxNLocator(5))
        ax1.zaxis.set_major_locator(plt.MaxNLocator(5))
               
        chi_sign = "+" if euler_char >= 0 else ""
        connectivity_state = L('high_connectivity') if euler_char < 0 else L('low_connectivity')
        ax1.set_title(f'\n'
                     f'{L("radius")} = {radius:.2f} μm | χ = {chi_sign}{euler_char}\n'
                     f'$\phi$ = {porosity_real * 100:.1f}% | {connectivity_state}', 
                     fontsize=FONT_SIZE_SMALL, pad=10, weight='normal')
        
        ax1.view_init(elev=30, azim=45)
        ax1.grid(True, alpha=0.2)
        
        # Panel 2: Función de conectividad
        ax2.plot(epc_results['radii'], epc_results['euler_chars'], 
                'lightgray', linewidth=1, alpha=0.3)
        
        ax2.plot(epc_results['radii'][:idx+1], epc_results['euler_chars'][:idx+1], 
                'b-', linewidth=3)
        
        point_color = 'blue' if euler_char < 0 else 'red'
        ax2.plot(radius, euler_char, 'o', color=point_color, markersize=15, 
                zorder=5, markeredgecolor='darkred', markeredgewidth=2)
        
        ax2.axhline(y=0, color='red', linestyle='--', linewidth=2.5, alpha=0.8)
        ax2.text(epc_results['radii'][-1]*0.95, 0, 'χ = 0', 
                verticalalignment='bottom', horizontalalignment='right',
                fontsize=FONT_SIZE_SMALL, color='red')
        
        if 'critical_radius' in epc_results and epc_results['critical_radius'] is not None:
            ax2.axvline(x=epc_results['critical_radius'], 
                       color='green', linestyle='--', linewidth=2.5,
                       label=f'{L("rc_topological")} = {epc_results["critical_radius"]:.2f} μm')
        
        if 'critical_radius_physical' in epc_results and epc_results['critical_radius_physical'] is not None:
            ax2.axvline(x=epc_results['critical_radius_physical'], 
                       color='red', linestyle=':', linewidth=2.5,
                       label=f'{L("rc_derivative")} = {epc_results["critical_radius_physical"]:.2f} μm')
        
        ax2.fill_between(epc_results['radii'][:idx+1], 0, epc_results['euler_chars'][:idx+1],
                        where=(epc_results['euler_chars'][:idx+1] < 0),
                        color='blue', alpha=0.1)
        ax2.fill_between(epc_results['radii'][:idx+1], 0, epc_results['euler_chars'][:idx+1],
                        where=(epc_results['euler_chars'][:idx+1] > 0),
                        color='red', alpha=0.1)
        
        ax2.set_xlim([0, epc_results['radii'][-1] * 1.05])
        ax2.set_ylim([euler_min-1, euler_max+1])
        ax2.set_xlabel(L('pore_radius'), fontsize=FONT_SIZE_MAIN)
        ax2.set_ylabel(L('euler_char'), fontsize=FONT_SIZE_MAIN)
        ax2.set_title(f'{L("connectivity_evolution")}\nχ = {euler_char}', 
                     fontsize=FONT_SIZE_TITLE)
        ax2.grid(True, alpha=0.3, which='both')
        ax2.tick_params(axis='both', which='major', labelsize=FONT_SIZE_MAIN)
        
        if (('critical_radius' in epc_results and epc_results['critical_radius'] is not None) or
            ('critical_radius_physical' in epc_results and epc_results['critical_radius_physical'] is not None)):
            ax2.legend(loc='best', fontsize=FONT_SIZE_LEGEND, framealpha=0.9)
        
        progress = (frame_idx / (n_frames - 1)) * 100 if n_frames > 1 else 0
        domain_info = f" | {L('domain')}: {domain_size_um[0]:.1f}×{domain_size_um[1]:.1f}×{domain_size_um[2]:.1f} μm³"
        fig.suptitle(f'{L("epc_3d_evolution")} {frame_idx+1}/{n_frames} | {L("progress")}: {progress:.0f}%', 
                    fontsize=FONT_SIZE_SUPTITLE, y=0.98)
        
        plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    
    print(L('generating_frames'))
    anim = FuncAnimation(fig, update, frames=n_frames, interval=1000//fps, repeat=True)
    
    filename = 'epc_evolution_3d_isometric.gif'
    print(L('saving_animation').format(filename))
    anim.save(filename, writer=PillowWriter(fps=fps))
    print(L('animation_saved'))
    print(f"{L('animated_domain')}: {domain_size_um[0]:.1f} × {domain_size_um[1]:.1f} × {domain_size_um[2]:.1f} μm³")
    
    plt.close(fig)
    
    if 'create_epc_static_comparison' in globals():
        create_epc_static_comparison(epc_results)
    
    return anim
```

```python
def main():
    """
    Pipeline completo con cálculo de EPC.
    """
    
    total_start = time.time()
    
    print("="*80)
    print(L('super_code_title'))
    print(L('corrected_version'))
    print("="*80)
    
    # PASO 1: Generar medio heterogéneo
    print(f"\n{L('step1')}")
    
    shape = (150, 150, 150)
    
    medium_info = create_heterogeneous_tight_rock(
        shape=shape,
        porosity_target=0.05,
        voxel_size_um=0.05,
        seed=42
    )
    
    # PASO 2: Análisis EPC
    print(f"\n{L('step2')}")
    
    epc_results = calculate_connectivity_function_enhanced(
        medium=medium_info['medium'],
        voxel_size_um=medium_info['voxel_size_um'],
        n_points=30
    )
    
    # PASO 3: Visualización
    print(f"\n{L('step3')}")
    
    try:
        create_epc_static_comparison(epc_results)
    except Exception as e:
        print(f"Error en visualización EPC: {e}")
        
    # PASO 8: Animación EPC 3D    
    print(f"\n{L('step8')}")
    
    try:
        anim_epc = create_animated_epc_evolution(epc_results, fps=2)
    except Exception as e:
        print(f"Error en animación EPC 3D: {e}")

    # RESUMEN FINAL
    total_time = time.time() - total_start
    
    print(f"\n{'='*80}")
    print(L('analysis_completed'))
    print(f"{'='*80}")
    print(f"{L('total_time')}: {total_time:.1f}s ({total_time/60:.1f} min)")
    
    print(f"\n{L('files_generated')}:")
    print(f"• {L('file_evolution')}")
```

```python
if __name__ == "__main__":
    main()
```
